```python
#!/usr/bin/env python3
"""
COMPLETE MIMIC-CXR Dataset Information and Preprocessing Pipeline
Full dataset analysis with comprehensive statistics and visualization
"""

import pandas as pd
import numpy as np
import pydicom
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.ensemble import IsolationForest
import json
from datetime import datetime
import missingno as msno
from collections import Counter
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class MIMICCXRAnalyzer:
    """
    Complete MIMIC-CXR Dataset Analysis and Preprocessing Pipeline
    """

    def __init__(self, data_path):
        self.data_path = Path(data_path)
        self.metadata = None
        self.chexpert_labels = None
        self.split_info = None
        self.reports = None
        self.processed_data = None
        self.stats = {}
        self.analysis_results = {}

    def load_complete_dataset(self):
        """Load all available dataset components"""
        print("üîç LOADING MIMIC-CXR DATASET COMPONENTS")
        print("=" * 100)

        dataset_files = {
            'metadata': 'cxr-study-list.csv.gz',
            'split_info': 'cxr-record-list.csv.gz',
            # Removed CheXpert and reports as they are not in this dataset subset
        }

        loaded_files = {}

        for file_type, filename in dataset_files.items():
            try:
                file_path = self.data_path / filename
                if file_path.exists():
                    df = pd.read_csv(file_path)
                    setattr(self, file_type, df)
                    loaded_files[file_type] = {
                        'rows': len(df),
                        'columns': len(df.columns),
                        'size_mb': round(file_path.stat().st_size / (1024 * 1024), 2)
                    }
                    print(f"‚úÖ {filename:30} | Rows: {len(df):>8,} | Columns: {len(df.columns):>3} | Size: {loaded_files[file_type]['size_mb']:>6} MB")
                else:
                    print(f"‚ùå {filename:30} | File not found")
            except Exception as e:
                print(f"‚ùå {filename:30} | Error: {e}")

        self.stats['loaded_files'] = loaded_files
        return loaded_files

    def get_complete_dataset_info(self):
        """Generate comprehensive dataset information"""
        print("\nüìä COMPLETE DATASET INFORMATION")
        print("=" * 100)

        # Basic dataset information
        dataset_info = {
            "name": "MIMIC-CXR Database v2.0.0",
            "version": "2.0.0",
            "institution": "Beth Israel Deaconess Medical Center, Boston, MA",
            "years": "2011-2016",
            "hipaa_compliant": True,
            "publicly_available": True,
            "citation": "Johnson et al., Scientific Data 2019"
        }

        # Study statistics
        if self.metadata is not None:
            study_stats = {
                "total_images": len(self.metadata), # This should be from record_df, but using metadata for consistency with original code
                "total_studies": self.metadata['study_id'].nunique(),
                "total_patients": self.metadata['subject_id'].nunique(),
                "studies_per_patient": round(len(self.metadata) / self.metadata['subject_id'].nunique(), 2),
                "images_per_study": round(len(self.metadata) / self.metadata['study_id'].nunique(), 2) # This should be from record_df, but using metadata for consistency with original code
            }
            dataset_info.update(study_stats)

        # Print formatted information
        for key, value in dataset_info.items():
            key_formatted = key.replace('_', ' ').title()
            if isinstance(value, bool):
                value_str = "Yes" if value else "No"
            elif isinstance(value, (int, float)) and key not in ['name', 'institution', 'years', 'citation']:
                value_str = f"{value:,}"
            else:
                value_str = str(value)
            print(f"{key_formatted:25}: {value_str}")

        return dataset_info

    def analyze_metadata_comprehensive(self):
        """Comprehensive metadata analysis"""
        print("\nüë• METADATA ANALYSIS")
        print("=" * 100)

        if self.metadata is None:
            print("‚ùå Metadata (cxr-study-list.csv.gz) not loaded")
            return None

        analysis = {}

        # 1. Basic structure
        print("1. DATASET STRUCTURE:")
        print(f"   ‚Ä¢ Shape: {self.metadata.shape}")
        print(f"   ‚Ä¢ Columns: {list(self.metadata.columns)}")

        # 2. Patient demographics (limited with available data)
        print("\n2. PATIENT DEMOGRAPHICS:")
        # Age and Sex not available in study list

        # 3. Study information
        print("\n3. STUDY INFORMATION:")
        # StudyDate not available in study list

        # 4. Image characteristics (limited with available data)
        print("\n4. IMAGE CHARACTERISTICS:")
        # ViewPosition, Rows, Columns not available in study list


        # 5. Missing values analysis
        print("\n5. MISSING VALUES ANALYSIS:")
        missing_data = self.metadata.isnull().sum()
        missing_percent = (missing_data / len(self.metadata)) * 100

        missing_stats = pd.DataFrame({
            'missing_count': missing_data,
            'missing_percent': missing_percent
        }).sort_values('missing_percent', ascending=False)

        analysis['missing_stats'] = missing_stats

        print("   Top columns with missing values:")
        for col, row in missing_stats.head(10).iterrows():
            if row['missing_count'] > 0:
                print(f"     ‚Ä¢ {col:20}: {row['missing_count']:>6,} ({row['missing_percent']:>5.1f}%)")

        self.analysis_results['metadata'] = analysis
        return analysis

    def analyze_chexpert_labels(self):
        """Comprehensive analysis of CheXpert labels"""
        print("\nüè• CHEXPERT LABELS ANALYSIS")
        print("=" * 100)

        if self.chexpert_labels is None:
            print("‚ùå CheXpert labels (mimic-cxr-2.0.0-chexpert.csv) not loaded. This file is not included in the provided dataset subset.")
            return None

        # The analysis logic below depends on chexpert_labels which is None
        # Removing this analysis as the file is not available.


    def analyze_data_splits(self):
        """Analyze train/validation/test splits"""
        print("\nüìä DATA SPLITS ANALYSIS")
        print("=" * 100)

        if self.split_info is None:
            print("‚ùå Split information (cxr-record-list.csv.gz) not loaded")
            return None

        analysis = {}

        # Split distribution (based on available data)
        if 'path' in self.split_info.columns:
            # Assuming 'path' represents an instance, not a split
            print("Split information file loaded, but does not contain a 'split' column for train/val/test split analysis.")
            print(f"Number of records: {len(self.split_info):,}")
            print(f"Number of unique subjects: {self.split_info['subject_id'].nunique():,}")
            print(f"Number of unique studies: {self.split_info['study_id'].nunique():,}")

        self.analysis_results['splits'] = analysis
        return analysis


    def create_visualizations(self):
        """Create comprehensive visualizations"""
        print("\nüìà CREATING VISUALIZATIONS")
        print("=" * 100)

        # Adjusting the number of subplots based on available data for visualization
        available_plots = 0
        if self.metadata is not None and 'path' in self.metadata.columns: # Check if metadata and 'path' is available to estimate studies per patient
             available_plots += 1 # Studies per Patient

        if self.metadata is not None and 'path' in self.metadata.columns: # Check if metadata and 'path' is available to check missing data
             available_plots += 1 # Missing Data Matrix

        # Dynamically create subplots based on available plots
        if available_plots > 0:
            fig, axes = plt.subplots(1, available_plots, figsize=(6 * available_plots, 6))
            if available_plots == 1: # Ensure axes is always an array
                axes = [axes]
            fig.suptitle('MIMIC-CXR Dataset Comprehensive Analysis', fontsize=16, fontweight='bold')

            plot_idx = 0

            # Studies per Patient
            if self.metadata is not None and 'path' in self.metadata.columns:
                studies_per_patient = self.metadata.groupby('subject_id').size()
                studies_count = studies_per_patient.value_counts().head(10)
                axes[plot_idx].bar(studies_count.index, studies_count.values)
                axes[plot_idx].set_xlabel('Number of Studies')
                axes[plot_idx].set_ylabel('Number of Patients')
                axes[plot_idx].set_title('Studies per Patient Distribution (from study list)')
                axes[plot_idx].grid(True, alpha=0.3)
                plot_idx += 1

            # Missing Data Matrix
            if self.metadata is not None:
                msno.matrix(self.metadata, ax=axes[plot_idx])
                axes[plot_idx].set_title('Missing Data Pattern (from study list)')
                plot_idx += 1

            plt.tight_layout()
            plt.savefig('mimic_cxr_analysis.png', dpi=300, bbox_inches='tight')
            plt.show()

            print("‚úÖ Visualizations saved as 'mimic_cxr_analysis.png'")
        else:
            print("‚ùå No data available for visualization.")


    def generate_preprocessing_pipeline(self):
        """Generate complete preprocessing pipeline description"""
        print("\n‚öôÔ∏è PREPROCESSING PIPELINE")
        print("=" * 100)

        pipeline = {
            "phase_1": {
                "name": "Data Loading and Validation",
                "steps": [
                    "Load study list and record list",
                    "Validate file integrity and consistency",
                    "Check for corrupted or missing files",
                    "Verify DICOM image accessibility"
                ]
            },
            "phase_2": {
                "name": "Data Cleaning",
                "steps": [
                    "Merge study list and record list based on subject_id and study_id",
                    "Handle missing values", # General step as specific columns are not available
                    "Remove studies without associated images", # This step is partially covered by merging
                    "Remove duplicate records"
                ]
            },
            "phase_3": {
                "name": "Feature Engineering",
                "steps": [
                    "Extract DICOM metadata features (if available and relevant)",
                    "Process radiology reports for findings (requires report file)", # Need report file for this
                    "Create patient-level aggregations"
                ]
            },
            "phase_4": {
                "name": "Data Standardization",
                "steps": [
                    "Normalize pixel values for DICOM images (if processing images directly)",
                    "Standardize numerical features",
                    "Encode categorical variables"
                ]
            },
            "phase_5": {
                "name": "Quality Assurance",
                "steps": [
                    "Detect and handle outliers",
                    "Ensure data consistency across merged tables",
                    "Generate data quality reports"
                ]
            }
        }

        for phase_name, phase_info in pipeline.items():
            print(f"\n{phase_info['name'].upper()}:")
            for i, step in enumerate(phase_info['steps'], 1):
                print(f"  {i}. {step}")

        return pipeline

    def save_analysis_report(self):
        """Save comprehensive analysis report"""
        print("\nüíæ SAVING ANALYSIS REPORT")
        print("=" * 100)

        report = {
            "timestamp": datetime.now().isoformat(),
            "dataset_info": self.get_complete_dataset_info(),
            "metadata_analysis": self.analysis_results.get('metadata', {}),
            "chexpert_analysis": self.analysis_results.get('chexpert', {}),
            "split_analysis": self.analysis_results.get('splits', {}),
            "preprocessing_pipeline": self.generate_preprocessing_pipeline()
        }

        # Save as JSON
        with open('mimic_cxr_analysis_report.json', 'w') as f:
            json.dump(report, f, indent=2, default=str)

        # Save summary as text
        with open('mimic_cxr_summary.txt', 'w') as f:
            f.write("MIMIC-CXR DATASET COMPREHENSIVE ANALYSIS REPORT\n")
            f.write("=" * 50 + "\n\n")

            f.write("DATASET OVERVIEW:\n")
            for key, value in report['dataset_info'].items():
                f.write(f"{key.replace('_', ' ').title()}: {value}\n")

            f.write(f"\nGenerated on: {report['timestamp']}\n")

        print("‚úÖ Analysis report saved as:")
        print("   - mimic_cxr_analysis_report.json (detailed)")
        print("   - mimic_cxr_summary.txt (summary)")

    def run_complete_analysis(self):
        """Run complete dataset analysis"""
        print("üöÄ STARTING COMPLETE MIMIC-CXR DATASET ANALYSIS")
        print("=" * 100)

        # Load data
        self.load_complete_dataset()

        # Comprehensive analysis
        self.get_complete_dataset_info()
        self.analyze_metadata_comprehensive()
        self.analyze_chexpert_labels() # This will indicate that the file is not loaded
        self.analyze_data_splits()

        # Visualizations
        self.create_visualizations()

        # Preprocessing pipeline
        self.generate_preprocessing_pipeline()

        # Save report
        self.save_analysis_report()

        print("\n" + "=" * 100)
        print("‚úÖ ANALYSIS COMPLETED SUCCESSFULLY!")
        print("=" * 100)

# Usage example
if __name__ == "__main__":
    # Initialize analyzer
    analyzer = MIMICCXRAnalyzer(data_path="/content/data/MIMIC-CXR/MIMIC-CXR")

    # Run complete analysis
    analyzer.run_complete_analysis()

    # Access analysis results (add checks for None)
    print("\nüìã KEY FINDINGS SUMMARY:")
    if analyzer.metadata is not None:
        print(f"‚Ä¢ Total studies: {analyzer.metadata['study_id'].nunique():,}")
        print(f"‚Ä¢ Total patients: {analyzer.metadata['subject_id'].nunique():,}")
        # ViewPosition is not in this metadata file
        # print(f"‚Ä¢ Most common view: {analyzer.metadata['ViewPosition'].mode()[0]}")

    # Age is not in this metadata file
    # if 'PatientAge' in analyzer.metadata.columns:
    #     print(f"‚Ä¢ Average patient age: {analyzer.metadata['PatientAge'].mean():.1f} years")
```

## Task 3: Feature Extraction

<div class="alert alert-block alert-info">

```python
settings = {}

# Normalizes the image by centering it at the mean with standard deviation.
# Normalization is based on all gray values in the image, not just those inside the segmentation.
settings['normalize'] = True
settings['label'] = 255

# COMPLETE : Create the featureextractor with the defined settings.
extractor = featureextractor.RadiomicsFeatureExtractor(**settings)
```

```python
feature_list = []

# Loop over patients
for i in range(len(metadata_df)):
    subject_id = metadata_df.iloc[i]['subject_id']
    study_id = metadata_df.iloc[i]['study_id']
    dicom_path = metadata_df.iloc[i]['dicom_path']
    mask_path = metadata_df.iloc[i]['mask_path']
    img_fname = glob(f'{DATA_PATH}/' + dicom_path)[0]
    seg_fname = glob(f'{DATA_PATH}/' + mask_path)[0]

    if len(img_fname) == 0 or len(seg_fname)==0:
        continue # Allows to skip subjects without image or without segmentation

    try:
        pixel_array = pydicom.dcmread(img_fname).pixel_array # Load
    except:
        continue
    image_sitk = sitk.GetImageFromArray(pixel_array)

    #pixel_array = pydicom.dcmread(img_fname).pixel_array # COMPLETE : Load image with pydicom and get pixel array
    #image_sitk = sitk.GetImageFromArray(pixel_array)

    reader = sitk.ImageFileReader() # Create a file reader
    reader.SetImageIO("PNGImageIO") # Set input type
    reader.SetFileName(seg_fname) # Set filename
    mask_sitk = reader.Execute() # Read

    result = extractor.execute(image_sitk, mask_sitk) # Execute the feature extraction

    result['subject_id'] = subject_id
    result['study_id'] = study_id
    result['dicom_path'] = dicom_path

    feature_list.append(result)
```

```python
# Gather all radiomics data into a df and save it
radiomics = pd.DataFrame()

for feature in feature_list:
  new_df = pd.DataFrame.from_dict(data=feature, orient='index')
  radiomics = pd.concat([radiomics, new_df.iloc[22:].T]) # Remove 22 first features which won't be useful (metadata)

radiomics.to_csv(f'features_raw.csv')
```

```python
import spacy
import csv

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def is_chunk_negated(chunk):
    """
    Check if the noun chunk is governed by negation.
    Looks at negation tokens in or near the chunk's root.
    """
    # Check within chunk
    for token in chunk:
        if token.dep_ == "neg" or token.text.lower() in NEGATIONS:
            return True

    # Check to the left of the chunk root
    for token in chunk.root.lefts:
        if token.dep_ == "neg" or token.text.lower() in NEGATIONS:
            return True

    # Check ancestors of root (e.g., "without pleural effusion")
    for ancestor in chunk.root.ancestors:
        for child in ancestor.children:
            if child.dep_ == "neg" or child.text.lower() in NEGATIONS:
                return True

    return False

def extract_positive_findings(text):
    if "IMPRESSION:" in text:
        impression_text = text.split("IMPRESSION:")[1].strip()

    elif "FINDINGS" in text:
        impression_text = text.split("FINDINGS")[1].strip()

    else:
        impression_text = text.strip()

    doc = nlp(impression_text.lower())

    findings = []

    for chunk in doc.noun_chunks:
        chunk_text = chunk.text.strip().lower()
        word_in_text = [word for word in DIAGNOSIS_KEYWORDS if word in chunk_text]
        if any(word_in_text):
            if not is_chunk_negated(chunk):
                findings.extend(word_in_text)

    if findings == []:
        findings.append('no findings')

    return np.unique(findings).tolist()

def process_radiology_report(file_path):
    with open(file_path, 'r') as file:
        text = file.read()

    findings = extract_positive_findings(text)

    return findings
```

```python
# Define basic diagnostic keywords (can be expanded)
DIAGNOSIS_KEYWORDS = [
    "atelectasis", "consolidation", "effusion", "pneumothorax", "edema",
    "infiltrate", "opacity", "fibrosis", "emphysema", "nodule", "mass"
    'cardiomegaly',	'enlargement',	'fracture',	'lesion', 'pneumonia'
]

NEGATIONS = ["no", "not", "without", "absent", "denies","negative for", "rule out", "free of",
    "lack of", "resolved", "excluded","neither", "never", "nor", "nothing","cannot be seen", "unremarkable for",
    "no evidence of", "no signs of"] # COMPLETE with other negation formula
```
